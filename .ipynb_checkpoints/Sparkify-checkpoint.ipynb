```python
# After Exploratory Data Analysis and defining churn, I define x features for my model pedicting potential users who will churn.
# These features are 
# 1.Registration duration (loyalty), 
# 2.Total number of songs listened ,
# 3.Average number songs listened per session,
# 4.The number of thumbs up
# 5.The number of thumbs down
# 6.The number of add friends
# 7.The number of add playlist
# 8.Total lenght of listening 
# 9.Gender
#10.Level - paid or free
#11.Churn Status / Label
```

```python
def registered_days(df): 
    """Function for Registration duration (days) feature
    
    Args:
        df : input dataframe for further processing

    Returns:
        df: a processed dataframes which only has userId and Register_Duration
    """
    # Calculate register duration from registration date to last page event times. 
    last_page_event_df =  df.groupBy('userId', 'Churn_Flag').agg(max('ts').alias('last_page_event'))
    registration_df = df.groupBy('userId').agg(max('registration').alias('registration'))
    # Join two spark dataframes
    df_register_duration = last_page_event_df.join(registration_df, on=['userID'], how='inner')
    # Convert last page event and registration dates to human readable formats
    convert_ts = udf(lambda x: datetime.datetime.fromtimestamp(x / 1000.0).strftime("%Y-%m-%d %H:%M:%S"))
    df_register_duration = df_register_duration.withColumn('page_event_time_human', convert_ts('last_page_event'))
    df_register_duration = df_register_duration.withColumn('registration_human', convert_ts('registration'))
    #  Obtain day difference of last page event and registration dates and assign into Registration Duration
    df_register_duration = df_register_duration.withColumn("Register_Duration", f.datediff(f.col('page_event_time_human'), f.col('registration_human')))     
    return df_register_duration.select("userId","Register_Duration")
```

```python
def total_songs(df):
    """Function for total number of songs listened feature
    
    Args:
        df : input dataframe for further processing

    Returns:
        df: a processed dataframes which only has userId and Listened_Total_Song
    """
    df.createOrReplaceTempView("Sparkify_df_clean")
    df_1 = spark.sql('''
        SELECT Churn_Flag, userID, sum(page_counts) AS Listened_Total_Song
        FROM
        (SELECT Churn_Flag, userId, count(page) AS page_counts
        FROM  Sparkify_df_clean
        WHERE page == "NextSong"
        GROUP BY Churn_Flag, userId)
        GROUP BY Churn_Flag, userID
''')
    return df_1.select("userId","Listened_Total_Song")
```

```python
def avg_songs_per_session(df):
    """Function for Average number songs listened per session feature
    
    Args:
        df : input dataframe for further processing

    Returns:
        df: a processed dataframes which only has userId and Avg_Listened_Song_Per_Session
    """
    df.createOrReplaceTempView("Sparkify_df_clean")
    df_1 = spark.sql('''
        SELECT Churn_Flag, userID, avg(page_counts) AS Avg_Listened_Song_Per_Session
        FROM
        (SELECT Churn_Flag, userId, sessionid, count(page) AS page_counts
        FROM  Sparkify_df_clean
        WHERE page == "NextSong"
        GROUP BY Churn_Flag, userId, sessionid)
        GROUP BY Churn_Flag, userID
''')
    return df_1.select("userId","Avg_Listened_Song_Per_Session")
```

```python
def thumbs_up(df):
    """Function for the number of thumbs up feature"""
    df = df \
        .select('userID','page') \
        .where(df.page == 'Thumbs Up') \
        .groupBy('userID') \
        .count() \
        .withColumnRenamed('count', 'Number_of_Thumbs_Up') 
    return df
```

```python
def thumbs_down(df):
    """Function for the number of thumbs down feature"""
    df = df \
        .select('userID','page') \
        .where(df.page == 'Thumbs Down') \
        .groupBy('userID') \
        .count() \
        .withColumnRenamed('count', 'Number_of_Thumbs_Down') 
    return df
```

```python
def addplaylist(df):
    """Function for the number of add friend feature"""
    df = df \
        .select('userID','page') \
        .where(df.page == 'Add to Playlist') \
        .groupBy('userID') \
        .count() \
        .withColumnRenamed('count', 'Number_of_AddPlaylist') 
    return df
```

```python
def addfriend (df):
    """Function for the number of add playlist feature"""
    df = df \
        .select('userID','page') \
        .where(df.page == 'Add Friend') \
        .groupBy('userID') \
        .count() \
        .withColumnRenamed('count', 'Number_of_AddFriend') 
    return df
```

```python
def total_lenght (df):
    """Function for total lenght of listening  feature"""
    df = df \
    .select('userID','length') \
    .groupBy('userID') \
    .sum() \
    .withColumnRenamed('sum(length)', 'length')
    return df
```

```python
def gender (df): 
    """Function for gender feature"""
    from pyspark.sql.types import IntegerType
    df = df.select("userId", "gender").dropDuplicates()
    df_1 = df.replace(['M', 'F'], ['1', '0'], 'gender')  
    df_2 = df_1.withColumn("gender",df_1["gender"].cast(IntegerType()))
    return df_2
```

```python
def account_level (df):
    """Function for account level (paid or free) feature"""
    from pyspark.sql.types import IntegerType
    df = df.select("userId", "level").dropDuplicates()
    df_1 = df.replace(['paid', 'free'], ['1', '0'], 'level')  
    df_2 = df_1.withColumn("level",df_1["level"].cast(IntegerType()))
    return df_2
```

```python
def churn_flag (df):
    """Function for Churn Status / Label feature"""
    # create churn label
    flag_churn_event = udf(lambda x: 1 if x == 'Cancellation Confirmation' else 0, IntegerType())
    df = df.withColumn('churn_event', flag_churn_event('page'))
    # label user who churned
    windowval = Window.partitionBy('userId')
    df = df.withColumn('churn', max('churn_event').over(windowval))
    
    df.createOrReplaceTempView("Sparkify_df_clean")
    df_1 = spark.sql('''
        SELECT DISTINCT userID, churn as Churn_Label
        FROM  Sparkify_df_clean
''')
    return df_1.select("userId","Churn_Label")
```

```python
def aggregate_features_data (df):
    """Aggregate all features data which will be used in modeling"""
    # call functions created before
    df1 = registered_days(df)
    df2 = total_songs(df)
    df3 = avg_songs_per_session(df)
    df4 = thumbs_up(df)
    df5 = thumbs_down(df)
    df6 = addplaylist(df)
    df7 = addfriend (df)
    df8 = total_lenght (df)
    df9 = gender (df)
    df10= account_level (df)
    df11= churn_flag (df)
    #join all dataframes and generate consolidated features data
    all_aggs_df = df1.join(df2, on='userId')\
    .join(df3, on='userId')\
    .join(df4, on='userId')\
    .join(df5, on='userId')\
    .join(df6, on='userId')\
    .join(df7, on='userId')\
    .join(df8, on='userId')\
    .join(df9, on='userId')\
    .join(df10, on='userId')\
    .join(df11, on='userId')
    
    joined_df = all_aggs_df.drop_duplicates()
    features_data = joined_df.drop('userId')
     
    return features_data
```

# Modeling
Split the full dataset into train, test, and validation sets. Test out several of the machine learning methods you learned. Evaluate the accuracy of the various models, tuning parameters as necessary. Determine your winning model based on test accuracy and report results on the validation set. Since the churned users are a fairly small subset, I suggest using F1 score as the metric to optimize.

For this step of my project, I will be building 3 base models that could possibly be used for predicting churn. Since churn is a simple churned/not-churned value, I can leverage classification models. The models I will be building are below: 

Logistic Regression Model -(https://spark.apache.org/docs/latest/api/python/pyspark.ml.html#pyspark.ml.classification.LogisticRegression) 

Random Forest Model -
(http://spark.apache.org/docs/latest/ml-classification-regression.html#random-forest-regression)

GBT Classifier Model - (https://spark.apache.org/docs/latest/api/python/pyspark.ml.html#pyspark.ml.classification.GBTClassifier) 

From these models I will be providing their results and evaluating their Accuracy and F1-Score. It is important for me to understand when my models are being as accurate as possible when determining when one of my users will possibly be churning and thus this is something that I will check against all models. However, Accuracy is not everything thus I must also make sure to not have a model that will potentially tell me a lot more users could possibly be churning when in reality they may not or there may be too many users showing possible churn. This could potentially lead to time wasted on users that are not churning and could mean extra costs for Sparkify. To balance this, I will be evaluating and optimizing the F1-Score which is a function of Precision and Recall to help give me the right balanced model.

```python
# Read model data from json file.
model_data = spark.read.json('df_final_model_data.json')
# change column name Churn_Label to label
model_data = model_data.withColumnRenamed("Churn_Label","label")
# Check columns and shape
model_data.printSchema()
model_data.toPandas().shape
```

```python
# Three ML algorithms will be used to build model and f1 score will be used to optimize.  These are
# 1. Logistic Regression, 2. Random Forest Classifier and 3. Gradient Boosting Trees
# First of vectorize numerical variables in model data, transform them for feature pipeline.
columns = []   
for field in model_data.schema.fields :
            columns.append(field.name)

columns.remove('label')
assembler = VectorAssembler(inputCols=columns, outputCol="num_features")
model_data = assembler.transform(model_data)

# using standard scaler
scaler = StandardScaler(inputCol="num_features", outputCol="features", withStd=True)
scalerModel = scaler.fit(model_data)
model_data = scalerModel.transform(model_data)
```

```python
#Split the full dataset into train, test, and validation sets.
train, rest = model_data.randomSplit([0.6, 0.4], seed=42)
validation, test = rest.randomSplit([0.5, 0.5], seed=42)
```

```python
# import time to calculate model training duration (seconds)
from time import time

# 1. Logistic Regression 

# initialize classifier, set evaluater and build paramGrid
lr = LogisticRegression(maxIter=10)
f1_evaluator = MulticlassClassificationEvaluator(metricName='f1')
paramGrid = ParamGridBuilder().build()
crossval_lr = CrossValidator(estimator=lr, evaluator=f1_evaluator, estimatorParamMaps=paramGrid,numFolds=3)

# Calculate time metric of model. 
start_time = time()
cvModel_lr = crossval_lr.fit(train)
end_time = time()
cvModel_lr.avgMetrics
seconds = end_time- start_time

results_lr = cvModel_lr.transform(validation)

evaluator = MulticlassClassificationEvaluator(predictionCol="prediction")

print('Logistic Regression Metrics:')
print('Accuracy of model is : {}'.format(evaluator.evaluate(results_lr, {evaluator.metricName: "accuracy"})))
print('F1 score of model is :{}'.format(evaluator.evaluate(results_lr, {evaluator.metricName: "f1"})))
print('The training process of model took {} seconds'.format(seconds))
```

```python
# 2. Random Forest Classifier 

# initialize classifier, set evaluater and build paramGrid
rf = RandomForestClassifier()
f1_evaluator = MulticlassClassificationEvaluator(metricName='f1')
paramGrid = ParamGridBuilder().build()
crossval_rf = CrossValidator(estimator=rf,estimatorParamMaps=paramGrid,evaluator=f1_evaluator,numFolds=3)

# Calculate time metric of model. 
start_time = time()
cvModel_rf = crossval_rf.fit(train)
end_time = time()
cvModel_rf.avgMetrics
seconds = end_time- start_time

results_rf = cvModel_rf.transform(validation)

evaluator = MulticlassClassificationEvaluator(predictionCol="prediction")

print('Random Forest Metrics:')
print('Accuracy of model is : {}'.format(evaluator.evaluate(results_rf, {evaluator.metricName: "accuracy"})))
print('F1 score of model is :{}'.format(evaluator.evaluate(results_rf, {evaluator.metricName: "f1"})))
print('The training process of model took {} seconds'.format(seconds))
```

```python
# 3. Gradient Boosting Trees

# initialize classifier, set evaluater and build paramGrid
gbt = GBTClassifier(maxIter=10,seed=42)
f1_evaluator = MulticlassClassificationEvaluator(metricName='f1')
paramGrid = ParamGridBuilder().build()
crossval_gbt = CrossValidator(estimator=gbt,estimatorParamMaps=paramGrid,evaluator=f1_evaluator,numFolds=3)

# Calculate time metric of model. 
start_time = time()
cvModel_gbt = crossval_gbt.fit(train)
end_time = time()
cvModel_gbt.avgMetrics
seconds = end_time- start_time

results_gbt = cvModel_gbt.transform(validation)

evaluator = MulticlassClassificationEvaluator(predictionCol="prediction")

print('Gradient Boosted Trees Metrics:')
print('Accuracy of model is : {}'.format(evaluator.evaluate(results_gbt, {evaluator.metricName: "accuracy"})))
print('F1 score of model is :{}'.format(evaluator.evaluate(results_gbt, {evaluator.metricName: "f1"})))
print('The training process of model took {} seconds'.format(seconds))
```

### Summary of Three Models:
1. Logistics Regression Model takes  4,5 training seconds with 0,78 accuracy and 0.71  f1-score
2. Random Forrest Model  takes 4,8 training seconds with 0,90 accuracy and 0.89 f1-score
3. Gradient Boosted Trees Model  takes 21,4 training seconds with 0,84 accuracy and 0.83 f1-score

****Why Random Forrest Model is best for our scenario:****

First of all we used f1-score to select our final model because our problem is classification one and f1-score can help us to find the balance between accuracy and recall. Higher the f1-score the more perfect our model will be as false negative and false positive will be less.

Here Random Forrest has best f1-score that's why I choose it for the next steps.


```python
# Optimizing Hyperparameters in Random Forest Classification
clf = RandomForestClassifier()
numTrees=[20,75]
maxDepth=[10,20]    
paramGrid = ParamGridBuilder().addGrid(clf.numTrees, numTrees).addGrid(clf.maxDepth, maxDepth).build()       
crossval = CrossValidator(estimator = Pipeline(stages=[clf]),
                         estimatorParamMaps = paramGrid,
                         evaluator = MulticlassClassificationEvaluator(metricName='f1'),
                         numFolds = 3)

cvModel_rf = crossval.fit(train)
predictions = cvModel_rf.transform(test)

evaluator = MulticlassClassificationEvaluator(metricName='f1')
f1_score = evaluator.evaluate(predictions.select(col('label'), col('prediction')))
print('The F1 score is {:.2%}'.format(f1_score)) 
   
bestPipeline = cvModel_rf.bestModel

print('Best parameters : max depth:{}, num Trees:{}'.format(bestPipeline.stages[0].getOrDefault('maxDepth'), bestPipeline.stages[0].getNumTrees))
```

```python
# Best Model: Random Forrest Model with max depth:10, num Trees:75 parameters
rf_best = RandomForestClassifier(numTrees=75, maxDepth=10)
rf_best_model = rf_best.fit(train)
result = rf_best_model.transform(test)
evaluator = MulticlassClassificationEvaluator(predictionCol="prediction")
print('Random Forrest Model - Test Metrics with best parameters:')
print('Accuracy of model is :  {}'.format(evaluator.evaluate(result, {evaluator.metricName: "accuracy"})))
print('F1 score of model is : {}'.format(evaluator.evaluate(result, {evaluator.metricName: "f1"})))
```
